```python
games = pd.read_csv('games.csv')
games.info()
```

```python
# Check Nan values
games.isna().sum()
```

```python
games['PTS_home'] = games['PTS_home'].fillna(games['PTS_home'].mean())
games['FG_PCT_home'] = games['FG_PCT_home'].fillna(games['FG_PCT_home'].mean())
games['FT_PCT_home'] = games['FT_PCT_home'].fillna(games['FT_PCT_home'].mean())
games['FG3_PCT_home'] = games['FG3_PCT_home'].fillna(games['FG3_PCT_home'].mean())
games['AST_home'] = games['AST_home'].fillna(games['AST_home'].mean())
games['REB_home'] = games['REB_home'].fillna(games['REB_home'].mean())
games['PTS_away'] = games['PTS_away'].fillna(games['PTS_away'].mean())
games['FG_PCT_away'] = games['FG_PCT_away'].fillna(games['FG_PCT_away'].mean())
games['FT_PCT_away'] = games['FT_PCT_away'].fillna(games['FT_PCT_away'].mean())
games['FG3_PCT_away'] = games['FG3_PCT_away'].fillna(games['FG3_PCT_away'].mean())
games['AST_away'] = games['AST_away'].fillna(games['AST_away'].mean())
games['REB_away'] = games['REB_away'].fillna(games['REB_away'].mean())
```

```python
# linear interpolate NaN values
# games = games.fillna(method='mean', limit=1)
games.drop(columns=['TEAM_ID_away'], inplace=True)
games.drop(columns=['TEAM_ID_home'], inplace=True)
```

```python
# check for duplicates
games.drop_duplicates(inplace=True)
```
